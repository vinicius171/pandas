```python
coluna_tipo= pokemon_df["primary_type"] # selecionar uma coluna especifica (tipo series, um array unidimensional)
coluna_tipo
```

```python
# e se eu quiser o tipo como o nme junto?
name_type_df = pokemon_df[["english_name", "primary_type", "secondary_type"]] #selecionar mais de uma coluna
name_type_df.sample(5)
```

```python

grass_df = pokemon_df[pokemon_df["primary_type"] == "grass"] #selecionar linhas com base em uma condição
grass_df.sample(5)
```

```python
#filtrar pokemon fraco contra tipo agua
pokemon_df[["english_name", "against_water", "against_fire", "against_grass"]] #selecionar mais de uma coluna
```

```python
weak_against_water_df = pokemon_df[pokemon_df["against_water"] > 2] #selecionar linhas com base em uma condição
weak_against_water_df[["english_name", "primary_type", "against_water"]]
```

```python
#filtrar pokémon fraco contra tipo água e terra ao mesmo tempo
weak_against_water_ground_df = pokemon_df[(pokemon_df["against_water"] > 1) & (pokemon_df["against_ground"] > 1)]

weak_against_water_ground_df[["english_name", "primary_type", "against_water", "against_ground"]]
```
